1.What does RGBA stand for?

Ans: RGBA stands for red,green,blue,alpha colors. It's value os a tuple of 4 intigers from 0 to 255.

2.From the Pillow module, how do you get the RGBA value of any images?

Ans: we need to import Image module and open the image and apply image.convert('RGBA') to convert the format from RGB to RGBA then use image.getdata() and we can print the RGBA value of each pixel.

```python
imgpath = input('please enter the image file path with quotes')
imgpath = eval(imgpath)
from PIL import Image
img = Image.open(imgpath)
rgba = img.convert('RGBA')
px_values = list(rgba.getdata())
px_values
```

4.Use your image and load in notebook then, How can you find out the width and height of an
Image object?

```python
imgpath = input('please enter the image file path with quotes')
imgpath = eval(imgpath)
from PIL import Image
img = Image.open(imgpath)
width = img.width
height = img.height
print('width of the image is ',width)
print('height of the image is ',height)
img
```

5.What method would you call to get Image object for a 100100 image, excluding the lower-left
quarter of it?

```python
'''By using this code we are making the RGB as 0,0,0 in the lower left quarter but to crop a image we can use
imageobj.crop((box tuple)) but here we need to exclude the lower left quarter of it, so i have made them black'''  
import numpy as np
imgpath = input('please enter the 100 x 100 image file path with quotes')
imgpath = eval(imgpath)
from PIL import Image
img = Image.open(imgpath)
img_arr = np.array(img)
img_arr[50:100,0:50] = (0,0,0)
img = Image.fromarray(img_arr)
img.show()
```
